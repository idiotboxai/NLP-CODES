```python
import pandas as pd
import re
import pickle
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 1. NLTK dictionaries 
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# 2. Load the dataset 
df = pd.read_pickle("News_dataset.pickle")
```

```python
# First 100 rows
df = df.head(100).copy()
```

```python
# 3. Setup tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# 4. Create a simple text cleaning function
def clean_my_text(text):
    text = str(text).lower()                
    text = re.sub(r'[^a-z ]', '', text)     
    
    words = text.split()                    
    
    clean_words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    
    return " ".join(clean_words)        
```

```python
# 5. Cleaning function to the text
df['Cleaned_Content'] = df['Content'].apply(clean_my_text)
```

```python
# 6. Label Encoding (Convert text categories to numbers)
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(df['Category'])
```

```python
# 7. TF-IDF Representation
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['Cleaned_Content'])
```

```python
# 8. Save Outputs
pickle.dump(tfidf_matrix, open("tfidf_matrix.pkl", "wb"))
pickle.dump(encoded_labels, open("encoded_labels.pkl", "wb"))
```

```python
# 9. Print results to show the examiner!
print("--- Final Results ---")
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
print("First 5 Encoded Labels:", encoded_labels[:5])
print("Success! Files are saved.")
```
