```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import normalize
from gensim.models import Word2Vec
```

```python
# 1. Load data 
df = pd.read_csv('data.csv')
# Combining 'Make' and 'Model' to create a simple sentence
text_data = (df['Make'] + ' ' + df['Model']).head(3).tolist()
print("--- Our Text Data ---")
print(text_data)
```

```python
# 2. Bag of Words (Counting words)
print("1. Bag of Words")
count_vec = CountVectorizer()
bow = count_vec.fit_transform(text_data)
print("Vocabulary:", count_vec.get_feature_names_out())
print(bow.toarray())
```

```python
# 3. Normalized Count (Converting to decimals/percentages)
print("2. Normalized Count")
bow_norm = normalize(bow, norm='l1')
print(bow_norm.toarray())
```

```python
# 4. TF-IDF (Finding important words)
print("3. TF-IDF")
tfidf_vec = TfidfVectorizer()
tfidf = tfidf_vec.fit_transform(text_data)
print(tfidf.toarray())
```

```python
# 5. Word2Vec Embeddings (Converting words to coordinates)
print("4. Word2Vec Embeddings")
# Word2Vec needs a list of words, so we split the sentences
words_list = [sentence.split() for sentence in text_data]
w2v_model = Word2Vec(words_list, min_count=1, vector_size=2) # size=2 means 2 numbers per word

# Let's check the numbers for 'BMW'
print("Numerical representation for 'BMW':", w2v_model.wv['BMW'])
```
