```python
import nltk
```

```python
from nltk.tokenize import (WhitespaceTokenizer, WordPunctTokenizer, TreebankWordTokenizer, TweetTokenizer, MWETokenizer)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "Hey! my name is Prathamesh. I am studying NLP and it's amazing. #learning #NLP"
```

```python
print("--- Tokenization Results ---")
```

```python
# 1. Whitespace
print("Whitespace:", WhitespaceTokenizer().tokenize(text))
```

```python
# 2. Punctuation-based
print("Punctuation:", WordPunctTokenizer().tokenize(text))
```

```python
# 3. Treebank
print("Treebank:", TreebankWordTokenizer().tokenize(text))
```

```python
# 4. Tweet
print("Tweet:", TweetTokenizer().tokenize(text))
```

```python
# 5. MWE (Multi-Word Expression)
mwe_tokenizer = MWETokenizer([('am', 'studying')]) 
print("MWE:", mwe_tokenizer.tokenize(text.split()), "\n")
```

```python
# Part B: Stemming
print("\n--- Stemming Results ---")
ps = PorterStemmer()
ss = SnowballStemmer("english")
word = "learning"
print(f"Porter Stemmer ({word}):", ps.stem(word))
print(f"Snowball Stemmer ({word}):", ss.stem(word))
```

```python
# Part C: Lemmatization
print("\n--- Lemmatization Results ---")
lemmatizer = WordNetLemmatizer()
print("Lemmatized (learning):", lemmatizer.lemmatize("learning", pos='v'))
```
